## Zeitreihen

Die Zeitreihenanalyse ist ein fundamentaler Bestandteil der Datenanalyse, der sich mit der Untersuchung von Datensätzen beschäftigt, die sequentielle Informationen über Zeitintervalle hinweg aufzeichnen. Diese Art der Analyse ermöglicht es, Muster, Trends und zyklische Variationen in den Daten zu identifizieren, zu verstehen und vorherzusagen. Die Bedeutung der Zeitreihenanalyse erstreckt sich über zahlreiche Anwendungsbereiche, von der Wettervorhersage und Finanzmarktanalyse bis hin zur Vorhersage von Verbrauchertrends und der Optimierung von Betriebsabläufen in der Industrie. Durch die Fähigkeit, zukünftige Ereignisse auf Basis historischer Daten vorherzusagen, ermöglicht die Zeitreihenanalyse Entscheidungsträgern in Unternehmen und Organisationen, informierte Strategien zu entwickeln und Risiken zu minimieren.[Link](https://gpt5.blog/zeitreihenanalyse-time-series-analysis/)

## Wozu dient die Zeitreihenanalyse?

Die Zeitreihenanalyse dient einer Vielzahl von Zwecken, darunter:

- Vorhersage zukünftiger Ereignisse durch Erkennen von Mustern in historischen Daten (z.B. Aktienkursvorhersagen).
- Identifikation von Trends, die Entscheidungsträgern helfen können, informierte Entscheidungen zu treffen (z.B. in der Wirtschaft).
- Verständnis der zyklischen Variationen von Zeitreihendaten (z.B. Saisonalität von Verkaufszahlen).

## Grundbegriffe der Zeitreihenanalyse

**Zeitreihe:** Eine Sequenz von Datenpunkten, die in aufeinanderfolgender zeitlicher Reihenfolge gemessen und aufgezeichnet werden.

Um die Zeitreihenanalyse effektiv zu nutzen, ist es wichtig, mit einigen grundlegenden Begriffen vertraut zu sein:

- **Datenpunkt:** Ein einzelner Wert oder eine Beobachtung in der Zeitreihe.
- **Zeitstempel:** Der spezifische Zeitpunkt, zu dem ein Datenpunkt erfasst wurde.
- **Trend:**  Die allgemeine Richtung, in die sich die Zeitreihe über einen längeren Zeitraum hinweg bewegt.

## Stationarität in der Zeitreihenanalyse verstehen

**Stationär:** Eine Zeitreihe ist stationär, wenn ihre statistischen Eigenschaften (wie Mittelwert und Varianz) über die Zeit konstant bleiben.

Die Stationarität ist ein Schlüsselkonzept in der Zeitreihenanalyse, da viele Analysemethoden davon ausgehen, dass die Zeitreihe stationär ist, oder Maßnahmen erfordern, um eine nicht-stationäre Zeitreihe in eine stationäre umzuwandeln. Es gibt zwei Haupttypen der Stationarität:

- **Schwach stationär:** Nur der Mittelwert und die Varianz der Zeitreihe bleiben über die Zeit konstant.
- **Stark stationär**: Alle statistischen Eigenschaften der Zeitreihe bleiben über die Zeit konstant.

## Verschiedene Arten von Zeitreihen

Es gibt verschiedene Arten von Zeitreihen, die auf spezifische Weise analysiert und interpretiert werden müssen:


Art der Zeitreihe

* **Trend Zeitreihe:**    Eine Zeitreihe, die eine langfristige Zunahme  oder Abnahme aufweist.

* **Saisonale Zeitreihe:**	        Eine Zeitreihe mit regelmäßigen Mustern, die sich in einem festen Zeitrahmen wiederholen, z.B. jährlich.

     
* **Zyklische Zeitreihe**	      Zeitreihen mit Mustern, die sich nicht nach einem festen Zeitrahmen wiederholen und oft mit der Wirtschaftslage in Verbindung stehen.


* **Rauschen**  Rauschen ist die zufällige Variation von Daten, die nicht durch andere Komponenten erklärt werden kann.



## Was ist Autokorrelation in einer Zeitreihe?

Autokorrelation ist das Maß für die Korrelation zwischen einer Zeitreihe und einer verschobenen Version ihrer selbst.

Mit anderen Worten: Es handelt sich um das Maß für die Ähnlichkeit zwischen Daten zu unterschiedlichen Zeitpunkten.

Wenn Sie beispielsweise die Temperatur von heute mit der Temperatur von gestern korrelieren, werden Sie feststellen, dass eine positive Korrelation besteht.

Mit anderen Worten: Wenn die Temperatur gestern hoch war, ist die Temperatur heute tendenziell auch hoch. Denken Sie daran, dass Korrelation keine Kausalität bedeutet: Die Temperatur von gestern hat nicht die Temperatur von heute verursacht, aber sie hängen zusammen.

In der Grafik sehen Sie, dass die höchste Autokorrelation mit dem Vortag besteht (Verzögerung 1) und im Vergleich zu weiter entfernten Tagen abnimmt.

Wenn die Daten völlig zufällig sind, das heißt, ein Wert keine Beziehung zum vorherigen Wert hat, ist die Autokorrelation Null. Ein Beispiel für eine Autokorrelation, die sehr nahe bei Null liegt, ist die Korrelation zwischen den Renditen aufeinanderfolgender Tage des Dax  Index.

Im Falle einer negativen Autokorrelation können wir an eine Reihe von körperlichen Übungen denken. ### 2.4 Autocorrelation

```python
# Plot the autocorrelation (ACF) of bike sharing demand
fig, ax = plt.subplots(figsize = (10,6))
plot_acf(df_sales_sum['sales'], lags = 100, ax = ax)
plt.show()
```

Das ist ein Autokorrelationsplot (ACF - Autocorrelation Function). Er zeigt, wie stark die aktuellen Werte in deiner Zeitreihe mit früheren Werten korrelieren.

PACF measures the correlation between a time series and its lag, after removing the effects of all shorter lags.

Purpose: To identify the direct relationship between an observation and its lag, without intermediate effects.

Main Difference between ACF and PACF:
- ACF: Measures overall correlation at each lag, including indirect effects.
- PACF: Measures direct correlation at each lag, excluding indirect effects.

**Was bedeutet diese Grafik?**
- Y-Achse: Die Korrelation (Werte zwischen -1 und 1).
- X-Achse: Die Lag-Zeit, also wie viele Zeitschritte zurück verglichen werden.
- Punkte mit vertikalen Linien: Die Korrelation für verschiedene Zeitabstände (Lags).
- Blau schattierter Bereich: Der Konfidenzbereich. Punkte außerhalb dieses Bereichs sind statistisch signifikant.

**Analyse Grafik:**

1- Hohe Anfangskorrelation: Der erste Punkt (Lag 0) ist immer 1, weil ein Wert mit sich selbst perfekt korreliert. Danach sieht man, dass die Autokorrelation langsam abnimmt.

2- Langsame Abnahme: Die Korrelation bleibt für viele Lags positiv und nimmt allmählich ab. Das zeigt, dass dein Zeitreihendatensatz eine starke zeitliche Abhängigkeit hat.

3- Saisonale Muster: Einige Peaks und Dips könnten auf saisonale Effekte hindeuten. Falls dein Datensatz tägliche Verkaufszahlen enthält, könnte es wöchentliche Muster geben.

4- Signifikante Autokorrelation: Mehrere Lags liegen außerhalb des Konfidenzintervalls, was bedeutet, dass frühere Werte immer noch einen Einfluss auf spätere Werte haben.

**Was kann man daraus ableiten?**

- Wenn die Korrelation langsam abnimmt, deutet das auf eine starke Trendkomponente hin.
- Falls Peaks in regelmäßigen Abständen auftreten, könnte das auf Saisonalität hinweisen.
- Falls du ein Vorhersagemodell baust, solltest du überlegen, saisonale oder autoregressive Komponenten (z. B. ARIMA oder Prophet mit saisonalen Effekten) einzubeziehen.

Ja! Das ist ein Partial Autocorrelation Function (PACF) Plot, der zeigt, wie stark ein Wert in deiner Zeitreihe direkt mit seinen vorherigen Werten korreliert, nachdem der Einfluss anderer Zeitpunkte entfernt wurde.

**Was bedeutet diese Grafik?**
- Y-Achse: Partial Autokorrelation (Werte zwischen -1 und 1).
- X-Achse: Lag (Verzögerung), also wie viele Zeitschritte zurück verglichen werden.
- Punkte mit vertikalen Linien: Die Korrelation für verschiedene Zeitabstände.
- Blau schattierter Bereich: Der Konfidenzbereich; Werte außerhalb sind statistisch signifikant.

**1- Starker erster Lag:**
- Der erste Lag (Lag 1) hat eine sehr hohe Korrelation (~0.9), was typisch für eine Zeitreihe mit starker Abhängigkeit ist.

**2. Schnelles Abfallen nach wenigen Lags:**

- Nach Lag 5 bis 10 wird die Korrelation schwächer und pendelt sich nahe 0 ein.
- Das bedeutet, dass die Verkaufszahlen vor allem von den letzten wenigen Tagen beeinflusst werden, aber nicht mehr stark von weiter zurückliegenden Werten.

**3. Signifikante Werte bis ca. Lag 10-15:**

- Einige Punkte sind über dem Konfidenzbereich, das deutet darauf hin, dass vergangene Werte bis zu diesen Lags noch Einfluss auf die aktuellen Werte haben.
- Danach flacht die PACF-Kurve ab, was darauf hinweist, dass ältere Werte keinen direkten Einfluss mehr haben.

**Was bedeutet das für dein Modell?**

- **AR-Prozess (AutoRegressive Modell):**
    - Da die PACF-Werte nach ca. Lag 10 abfallen, könnte ein AR(10)-Modell eine gute Wahl sein.
    - Das bedeutet, dass man für eine Vorhersage etwa die letzten 10 Tage berücksichtigen sollte.


- **Vergleich mit ACF-Plot:**
   - Falls dein ACF-Plot langsam abnimmt, aber PACF nach einigen Lags      schnell abfällt, könnte eine ARIMA-Modellierung (AutoRegressive Integrated Moving Average) sinnvoll sein.

**Fazit**
- Die Zeitreihe zeigt eine starke kurzfristige Abhängigkeit.
- Die ersten 10-15 Lags haben noch einen Einfluss, danach wird die Korrelation sehr gering.
- **Falls** du ein ARIMA-Modell verwenden möchtest, könnte ein AR(10) oder AR(15) als erster Testwert sinnvoll sein.

O ADF é um dos testes estatísticos mais usados para TS. Suas hipóteses nula e alternativa são:

## 3.2 Teste ADF

Diferente de alguns projetos que vemos com o Prophet, vamos fazer o teste para descobrir se nossa série é estacionária. O teste utilizado será o ***Augmented Dickey Fuller (ADF)***.

Como bem disse o Carlos:
> Normalmente, rejeitamos $H_0$ quando o *p-value* é menor ou igual a um determinado nível de significância. Normalmente, esse nível de significância é igual a 0.01 (1%), 0.05 (5%) ou até mesmo 0.1 (10%).

> Quanto menor o *p-value*, mais evidências contra a hipótese nula você tem. Por exemplo, caso seja usado $\text{p-value} = 0.05$, rejeitamos a $H_0$ com um nível de confiança de 95%.

> Caso o valor do teste fique acima do *threshold*, falharemos em rejeitar a hipótese nula. **Atenção, isso não significa que $H_0$ é verdadeira.** Significa apenas que o experimento não encontrou evidências para rejeitá-la.

## 3.4 Conclusões - Série Não-Estacionária
Tudo isso parece lindo e perfeito, mas como saber se nossas previsões estão boas de verdade?

Para isso utilizaremos as seguintes métricas:

* Mean Absolute Percentage Error - Mostra quanto as nossas previsões diferem do valor real em porcentagem.
* Mean Absolute Error - Valor absoluto do nosso erro na previsão em relação à série real.

Abaixo temos uma função para calcular o MAPE e MAE da nossa série.

```python
# definindo a função
def make_comparison_dataframe(historical, forecast):
  return forecast.set_index('ds')[['yhat', 'yhat_lower',
                                'yhat_upper']].join(historical.set_index('ds'))
cmp_df = make_comparison_dataframe(df_ts, forecast)

def calculate_forecast_errors(df_ts, prediction_size):

  df = df_ts.copy()
  df['e'] = df['y'] - df['yhat']
  df['p'] = 100 * df['e'] / df['y']
  predicted_part = df[-prediction_size:]

  error_mean = lambda error_name: np.mean(np.abs(predicted_part[error_name]))

  return{'MAPE': error_mean('p'), 'MAE': error_mean('e')}

# imprimindo o MAPE e MAE
for err_name, err_value in calculate_forecast_errors(cmp_df, prediction_size).items():
  print(err_name, err_value)
```

```python
import optuna
import numpy as np
import pandas as pd
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics

# ---------------------------
# 1. Datenaufbereitung und Aufteilung
# ---------------------------
# Annahme: df_log ist Ihr DataFrame mit den Spalten 'ds' (Datum) und 'y' (Zielvariable)
# Chronologische Sortierung
df_log = df_log.sort_values(by='ds').reset_index(drop=True)

# Aufteilung in 70% Training, 15% Validierung, 15% Test
n = len(df_log)
train_end = int(n * 0.7)
val_end = int(n * 0.85)

train_df = df_log.iloc[:train_end].copy()
val_df   = df_log.iloc[train_end:val_end].copy()
test_df  = df_log.iloc[val_end:].copy()

# Indizes zurücksetzen
train_df.reset_index(drop=True, inplace=True)
val_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)

print("Datenaufteilung:")
print("Trainingsdaten:", train_df.shape)
print("Validierungsdaten:", val_df.shape)
print("Testdaten:", test_df.shape)
```

```python
# ---------------------------
# 2. Hyperparameter-Tuning mit Optuna (auf Trainingsdaten mit Cross-Validation)
# ---------------------------
def objective(trial):
    # Definition des Parameterraums
    params = {
        'changepoint_prior_scale': trial.suggest_float('changepoint_prior_scale', 0.001, 0.1, log=True),
        'seasonality_prior_scale': trial.suggest_float('seasonality_prior_scale', 1.0, 15.0),
        'holidays_prior_scale': trial.suggest_float('holidays_prior_scale', 1.0, 10.0),
        'seasonality_mode': trial.suggest_categorical('seasonality_mode', ['additive', 'multiplicative']),
        'n_changepoints': trial.suggest_int('n_changepoints', 5, 25)
    }

    # Prophet-Modell initialisieren und trainieren (nur auf Trainingsdaten)
    model = Prophet(yearly_seasonality=True, weekly_seasonality=True, **params)
    model.add_country_holidays(country_name='US')
    model.fit(train_df)

    # Cross-Validation durchführen:
    # initial: Zeitraum der ersten Trainingsphase,
    # period: Schrittweite, horizon: Forecast-Horizont (hier 30 Tage)
    df_cv = cross_validation(model, initial='665 days', period='15 days', horizon='30 days', parallel="processes")
    df_metrics = performance_metrics(df_cv, metrics=['mape', 'mae'])

    avg_mape = df_metrics['mape'].mean()
    avg_mae = df_metrics['mae'].mean()
    trial.set_user_attr('avg_mae', avg_mae)

    return avg_mape  # Ziel: Minimierung des MAPE

# Optuna-Studie erstellen und optimieren
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=30)

# Beste Hyperparameter extrahieren
#best_params = study.best_params
#print("\n🏆 Beste Hyperparameter:", best_params)
#print("✅ Bestes Validierungs-MAPE (CV auf Training):", study.best_value)
```

```python
# Beste Hyperparameter extrahieren
best_params = study.best_params
print("\n🏆 Beste Hyperparameter:", best_params)
print("✅ Bestes Validierungs-MAPE (CV auf Training):", study.best_value)
```

```python
# ---------------------------
# 3. Finales Modelltraining und Evaluierung auf Testdaten
# ---------------------------
# Finales Training: Kombinieren von Trainings- und Validierungsdaten
final_train = pd.concat([train_df, val_df]).reset_index(drop=True)

final_model = Prophet(yearly_seasonality=True, weekly_seasonality=True, **best_params)
final_model.add_country_holidays(country_name='US')
final_model.fit(final_train)

# Prognose für den Testzeitraum:
# Berechnen der Anzahl Tage im Testdatensatz
n_test = len(test_df)
future_test = final_model.make_future_dataframe(periods=n_test, freq='D')
forecast_test = final_model.predict(future_test)

# Vorhersagen mit Testdaten zusammenführen
cmp_df_test = forecast_test.set_index('ds')[['yhat']].join(test_df.set_index('ds'))

# Funktion zur Fehlerberechnung (MAPE & MAE)
def calculate_forecast_errors(df):
    df = df.dropna(subset=['y'])
    df['e'] = df['y'] - df['yhat']
    df['p'] = 100 * df['e'] / df['y']
    return {'MAPE': np.mean(np.abs(df['p'])), 'MAE': np.mean(np.abs(df['e']))}

test_errors = calculate_forecast_errors(cmp_df_test)

print("\n🚀 Test-Ergebnisse:")
print("📌 Test MAPE:", round(test_errors['MAPE'],3),"%")
print("📌 Test MAE:", round(test_errors['MAE'],3))

```
